```python
import matplotlib.pyplot as plt
import numpy as np
```

# Plot

```python
hours = [1,2,3,4,5,6,7,8]
visitors = [50,60,70,80,90,100,110,120]

plt.plot(hours, visitors,
         color="Purple",
         marker = "+",
         linestyle=":",
         linewidth=1,
         label="Visitors")

plt.title("Website Visitors Per Hour")
plt.xlabel("Hour Of Day")
plt.ylabel("Number Of Visitors")
plt.legend()

plt.grid(True)
plt.show()
```

# Bar **Chart**

```python
product = ["lappy", "Mobile", "Tab", "Headphones", "Watch"]
sales = [250, 400, 390, 1001, 740]

plt.bar(product, sales, color = "orange", edgecolor = "purple", alpha=1)
plt.title("Product Sales Comparision")
plt.xlabel("Products")
plt.ylabel("Units Sold")
plt.show()
```

# **Histogram**

```python
marks = [45,55,56,76,56,66,78,87,65,54,43,45,63,53,27,91,69,59,37,40]

plt.hist(marks, bins=10, color="Skyblue", edgecolor="indigo")
plt.title("Distribution Of Student Marks")
plt.xlabel("Marks Range")
plt.ylabel("Number Of Students")

plt.show()
```

# Scatter **Plot**

```python
ad_spend = [2,4,5,7,8,9,10,11,12]
sales = [20,25,28,29,30,35,37,39,46]

sizes = [200,250,280,290,350,390,400,480,500]

colours = sales
plt.scatter(ad_spend, sales,
            s=sizes,
            c=colours,
            cmap='viridis',
            alpha=0.8,
            edgecolors='black',
            linewidths=1.2
            )

plt.title("Advertising Spend vs Sales (with Color & Sice)", fontsize=14, fontweight="bold")
plt.xlabel("Ad Spend ($1000s)", fontsize=12)
plt.ylabel("Sales ($1000s)", fontsize=12)
plt.colorbar(label = "Sales Value")
plt.grid(True, linestyle="--", alpha=1)
plt.show()
```

# **Pie Chart**

```python

brands=['Apple','Samsung','Oppo','Realme','Others']
market_share=[30,25,20,10,15]
plt.pie(market_share,labels=brands,autopct='%1.2f%%',startangle=90,explode=(0.1,0,0,0,0),shadow=True)
plt.title('Mobile Market Share')
plt.show()
```

# **Box Plot**

```python

scores=[
    [78,85,90,88,76,95,9],
    [65,70,68,72,75,80,85,78],
    [90,92,94,96,88,91,89,93],
    [62,60,61,90,65,63,67,61],
    [70,85,8,50,92,91,100,89]
]
plt.boxplot(scores,patch_artist=True,notch=False,vert=True,boxprops=dict(facecolor='blue',color='black'),labels=['A','B','C','D','E'])
plt.title('Exam Scores Distribution Across Classes')
plt.xlabel('Classes')
plt.ylabel('Scores')
plt.grid(True,linestyle=':',alpha=0.7)
plt.show()
```
